```python
shapefile_path = "Shapefiles/Tomato Farm Shapefile/tomato_farm_dharmendra.shp"
import geopandas as gpd
import rasterio
```

```python
def clipper(raster_path,vector_path,epsg_code):
    
    #Open vector file for the geometry
    shapefile = gpd.read_file(vector_path)
    df = shapefile.to_crs(epsg=int(epsg_code))
    geometry = df.iloc[0]['geometry']
    
    #Open raster file for clipping
    with rasterio.open(raster_path) as src:
        try:
            out_image, out_transform = rasterio.mask.mask(src, [geometry] , crop=True, filled=True)
            out_meta = src.meta
        except error:
            print('error clipping')           
    
    return out_image
```

# Data prepared for 2019

```python
for item in data_path[1:]:
    print(item)
    #Get the bands
    bands = list(pathlib.Path(item).glob("*"))
    print(bands)
    
    meta_list = bands[0].stem.split("_")
    print(meta_list[2][:4])
    
    #Clip each band and store the clip file in output folder
    
    red_arr = clipper(bands[0],shapefile_path,32643)
    
    nir_arr = clipper(bands[1],shapefile_path, 32643)
    
    ndvi = es.normalized_diff(nir_arr, red_arr)
    
    #ndvi = (nir_arr - red_arr ) / (nir_arr + red_arr)
    
    #print(type(ndvi), ndvi)
    
    print(ndvi.mean(),ndvi.max())
    
    data_to_write = [5,meta_list[2],meta_list[2][:4],round(ndvi.mean(),4),round(ndvi.max(),4),round(ndvi.min(),4),0,0,0,'Tomato']
    print(f" Writing data : {data_to_write}")
    
    with open("training_data.csv", 'a') as csvfile:
        # creating a csv writer object 
        csvwriter = csv.writer(csvfile) 
        
        # writing the fields 
        csvwriter.writerow(data_to_write)
        
        print("data written succesfully")
        
        
       
            
    
    
```
